# 머신러닝이란?
**머신러닝(Machine learning)이란?** "기계가 패턴을 학습하여 자동화하는 알고리즘" 이라고 정의할 수 있다.\
여기서 **알고리즘(Algorithms)은** "어떠한 문제를 해결하기 위한 일련의 절차나 방법"으로 정의할 수 있다.

- 머신러닝의 사용 사례
    - 구매추천
    - 번역
    - 자율주행차
    - 챗봇

---

### 인공지능, 머신러닝, 딥러닝


즉, 인공지능 > 머신러닝 > 딥러닝


##### 1. 인공지능 (Artificial Intelligence, AI)
> 인간의 사고, 판단, 학습, 문제 해결 등을 기계가 모방하게 하는 기술

- **핵심 개념**: ‘기계에게 *지능*을 부여한다’  
- **특징**:
  - 상황에 따라 스스로 판단하고 결정할 수 있음
  - 규칙 기반부터 학습 기반까지 모두 포함
- **예시**:
  - 체스 게임 AI
  - 스마트 스피커(Siri, Bixby)
  - 간단한 규칙 기반 챗봇


##### 2. 머신러닝 (Machine Learning, ML)
> 기계가 **데이터를 통해 학습**하고, **스스로 규칙을 만드는** 인공지능의 한 분야

- **핵심 개념**: 사람이 일일이 규칙을 만들지 않아도, *데이터를 통해 기계가 학습*
- **종류**:
  - 지도학습: 입력-출력 쌍이 존재 (예: 질병 예측)
  - 비지도학습: 군집화, 차원 축소 등
  - 강화학습: 보상을 통해 스스로 행동을 학습
- **예시**:
  - 스팸 이메일 분류기
  - 유전자 발현 데이터를 활용한 암 분류 모델


##### 딥러닝 (Deep Learning, DL)
> **인공신경망**을 활용해 **복잡한 패턴**을 학습하는 머신러닝의 하위 분야

- **핵심 개념**: 뇌의 신경망처럼 작동하는 다층 구조의 인공 신경망
- **특징**:
  - 이미지, 음성, 자연어 등 복잡하고 비정형적인 데이터 처리에 강함
- **예시**:
  - 얼굴 인식
  - 자율주행 자동차 비전 시스템
  - ChatGPT 같은 생성형 언어 모델


# 머신러닝의 학습 프로세스
머신러닝을 이해하기 위해 가장 먼저 알아야 하는 것은 바로 머신러닝의 기본 모형이다. \
머신러닝을 간단히 설명하자면

$y = f(x)$

- $y$ : 머신러닝 알고리즘에 의해 산출되는 예측치
- $x$ : 가지고 있는 데이터
- $f()$ : 함수로 머신러닝에서는 모델

즉, 가지고 있는 데이터를 우리가 학습하여 생성한 함수에 넣으면 그 결과로 어떤 문제에 대한 예측치를 생성하는 것


### 머신러닝 프로세스



## 머신러닝의 종류
머신러닝을 기본적으로 분류하면 지도학습과 비지도학습으로 나눌 수 있다.
- 지도학습(Supervised learning) : **문제와 답을** 함께 학습함으로써 미지의 문제에 대한 올바른 답을 예측하는 학습 / input ( Data + Label )
    - 회귀(Regression) : **연속형** 값인 y의 특징을 찾아 데이터 x를 사용하여 y 값을 예측
        - 예시 : 주식 가격 예측, 암 환자의 생존 기간(시간) 예측
    - 분류(Classification) **이산형** 값인 y의 특징을 찾아 데이터 x를 사용하여 y 값을 예측
        - 예시 : 암과 정상 예측, 고양이와 강아지 예측
- 비지도학습(Unsupervised learning) : 조력자의 **도움 없이(닶 없이)** 컴퓨터 스스로 학습하는 형태 / input ( Data )
    - 군집(Clustering) : x 데이터의 특징이 유사한 값들의 모임을 군집으로 표현하는 기법
- 강화학습(Reinforcement learning) : 컴퓨터가 세상에 존재하는 규칙을 스스로 시뮬레이션하면서 게임처럼 규칙을 학습하는 것
    - 예시 : 알파고
